- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
import numpy as np
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
estimators = [('hc_ward', AgglomerativeClustering(n_clusters=3, linkage='ward')),
            ('hc_complete', AgglomerativeClustering(n_clusters=3, linkage='complete')),
            ('hc_average', AgglomerativeClustering(n_clusters=3, linkage='average')),
            ('hc_single', AgglomerativeClustering(n_clusters=3, linkage='single')),
            ]
```

```python
fignum = 1

title = ['ward', 'complete', 'average', 'single']
for names, est in estimators:
    
    est.fit(X)
    label = est.labels_
    
    fig = plt.figure(fignum)
    ax = Axes3D(fig)
    ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=label.astype(np.float), edgecolor='k')
    
    ax.set_title(title[fignum-1])
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    
    fignum+=1
fig = plt.figure(fignum)
ax = Axes3D(fig)

for name, label in [('cls0', 0),
                    ('cls1', 1),
                    ('cls2', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean(),
              X[y == label, 2].mean(),
              name,
              horizontalalignment='center')
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y.astype(np.float), edgecolor='k')
ax.set_title('Ground Truth')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
```
