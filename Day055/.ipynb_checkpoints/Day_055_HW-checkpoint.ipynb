# 作業

- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
X
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
estimators = [('K_means_7', KMeans(n_clusters=7)), ('K_means_5', KMeans(n_clusters=5)), ('K_means_3', KMeans(n_clusters=3))]
```

```python
fignum = 1
titles = ['7 cluster', '5 cluster', '3 cluster']
for name, est in estimators:
    fig = plt.figure(fignum)
    ax = Axes3D(fig)
    est.fit(X)
    labels = est.labels_
    ax.scatter(X[:, 3], X[:, 0], X[:, 2],
               c=labels.astype(np.float), edgecolor='k')
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1
fig = plt.figure(fignum)
ax = Axes3D(fig)
for name, label in [('cls0', 0),
                   ('cls1', 1),
                   ('cls2', 2)]:
    ax.text3D(X[y==label, 3].mean(), X[y==label, 0].mean(), X[y==label, 2].mean(), name)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=y, edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.dist = 12
ax.set_title('Ground Truth')
```

```python
estimators = [('K_means_7', KMeans(n_clusters=7, init='random')), ('K_means_5', KMeans(n_clusters=5, init='random')), ('K_means_3', KMeans(n_clusters=3, init='random'))]
```
